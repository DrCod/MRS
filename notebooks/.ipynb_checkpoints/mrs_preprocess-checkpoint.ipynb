```python
import os
import sys
import numpy as np
import pandas as pd
import glob
```

```python
#path to water suppressed and unsuppressed fid signals
PATH_WATER ="../datasets_text/dataset"
#path to metabolite basis
PATH_METABOLITES_BASIS_SET ="../"
```

```python
def get_water_suppressed_datasets(num):
    data_set_name=PATH_WATER + str(num)
    return data_set_name + ".txt"
```

```python
get_water_suppressed_datasets(1)
```

```python
get_water_suppressed_datasets(25)
```

```python
def get_metabolite_basis():
    basis_set_list =["Ace","Ala","Asc","Asp","Cr","GABA","Glc","Gln","Glu","Gly","Lac","NAA","NAAG","Ins","GPC","GSH","Mac","PCho","PCr","PE","Tau"]
    basis_paths =[bm + ".txt" for bm in basis_set_list]
    return [PATH_METABOLITES_BASIS_SET + l for l in basis_paths]
```

```python
get_metabolite_basis()
```

```python
def load_all_datasets(num):
    x_list =[get_water_suppressed_datasets(i) for i in range(num)]
    return x_list[1:]
```

```python
load_all_datasets(29)
```

```python
real_ds =[]
imag_ds =[]
water_real_ds=[]
water_imag_ds=[]

with open("../datasets_text/dataset1.txt") as f:
        for line in f:
            line =line.strip()
            if line:
                data =line.split()
                ds_real = data[0]
                ds_imag =data[1]
                ds_water_real = data[2]
                ds_water_imag = data[3]
                real_ds.append(ds_real)
                imag_ds.append(ds_imag)
                water_real_ds.append(ds_water_real)
                water_imag_ds.append(ds_water_imag)
```
